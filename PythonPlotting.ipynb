### Plotting in Python using matplotlib.pyplot

```python
import numpy as np

import matplotlib.pyplot as plt
import pandas as pd
```

```python
# For compatibility across multiple platforms
path_of_the_file = "C:/Users/dadidelux/Desktop/DSP Training Files/DataSets/Cities.csv"
cities = pd.read_csv(path_of_the_file, encoding='utf-8')
path_of_the_file = "C:/Users/dadidelux/Desktop/DSP Training Files/DataSets/Countries.csv"
countries = pd.read_csv(path_of_the_file, encoding='utf-8')
```

```python
# Introduction to scatterplots
x = [1, 2, 4, 6, 9]
y = [4, 5, 2, 7, 5]
plt.scatter(x, y)
plt.show()
```

```python
# Scatterplot of temperature (y) versus latitude (x)
plt.xlabel('latitude')
plt.ylabel('temperature')
plt.scatter(cities['latitude'].astype(float), cities['temperature'].astype(float))
plt.show()
```

```python
# Scatterplot of latitude (y) versus longitude (x) colored by temperature
x = cities.longitude.tolist()
y = cities.latitude.tolist()
colors = []
for index in range(len(x)):
    if   x[index] < 8:
        colors.append('blue')
    elif y[index] > 50:
        colors.append('red')
    else: colors.append('yellow')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.scatter(x, y, c=colors)
plt.show()
```

```python
# Introduction to bar charts
bars = ['Bar1', 'Bar2', 'Bar3']
heights = [2, 6, 4]
plt.xticks([1,2,3], bars)
plt.bar([1,2,3], heights, align='center')
plt.show()
```

```python
# Bar chart showing populations of countries with 'ia' in their name
bars = countries.country.tolist()
heights = countries.population.tolist()
numbars = len(bars)
# range(numbars) in following command yields [0,1,...,numbars-1]
plt.xticks(range(numbars), bars, rotation='vertical')
plt.bar(range(numbars), heights, align='center')
plt.xlabel('country')
plt.ylabel('population')
plt.show()
```

```python
# Introduction to pie charts
sizes = [10, 20, 30]
lbls = ['small', 'medium', 'large']
plt.pie(sizes, labels=lbls, autopct='%1.1f%%')
plt.show()
```

```python
# Pie chart showing number of EU countries versus non-EU countries

numEU = (countries['EU'] == 'yes').tolist()
numNotEU = (countries['EU'] == 'no').tolist()

sizes = [len(numEU), len(numNotEU)]
lbls = ['EU', 'not EaU']
plt.pie(sizes, labels=lbls, autopct='%1.1f%%')
plt.show()
```
